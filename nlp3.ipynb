```python
def getlabels(doc1):
  processed_article = re.sub('[^A-Z]', ' ', doc1 )
  processed_article = re.sub(r'\s+', ' ', processed_article)
  lb = nltk.sent_tokenize(processed_article)
  lab = [nltk.word_tokenize(sent) for sent in lb]
  print(lab)
  labels=[]
  for i in range(len(lab)):
    for j in range(len(lab[i])):
      if (lab[i][j]=="LOC"):
        labels.append(lab[i][j])
      elif(lab[i][j]=="MIS"):
        labels.append(lab[i][j])
      elif(lab[i][j]=="PER"):
        labels.append(lab[i][j])
      elif(lab[i][j]=="ORG"):
        labels.append(lab[i][j])
      elif(lab[i][j]=="O"):
        labels.append(lab[i][j])
  
  return(labels)
```

```python
from gensim.models import KeyedVectors
# Load vectors directly from the file
model = KeyedVectors.load_word2vec_format('/content/gdrive/MyDrive/model.bin', binary=True)
```

```python
w2v_embeddings_index={}
for word in model.wv.vocab:
  w2v_embeddings_index[word] = model[word]
print('Loaded %s word vectors.'% len(w2v_embeddings_index))
```

```python
# create a weight matrix for words in training docs
embedding_matrix = np.zeros((vocab_size, 100))
for word, i in t.word_index.items():
    embedding_vector = w2v_embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
print("Embedding Matrix shape:", embedding_matrix.shape)
```

```python
import tensorflow as tf
embedding_layer = tf.keras.layers.Embedding(vocab_size, 100, weights=[embedding_matrix], input_length=64, trainable=False)
```

```python
# define model
input_placeholder= tf.keras.Input(shape=(64,), dtype='int32')
input_embedding = embedding_layer(input_placeholder)
lstm= tf.keras.layers.LSTM(units=20, activation='relu')(input_embedding)
preds = tf.keras.layers.Dense(921, activation='sigmoid', name = "activation")(lstm)
modellstm = tf.keras.models.Model(inputs=input_placeholder, outputs=preds)
# compile the model
modellstm.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.Adam(lr=0.001), metrics=['accuracy'])
# summarize the model
print(modellstm.summary())
print('\n # Fit model on training data')
# fit the model
history= modellstm.fit(X_train, y_train, epochs=1000)
```

```python
# evaluate the model
loss, accuracy = modellstm.evaluate(X_train, y_train, verbose=0)
print('Accuracy: %f' % (accuracy*100))
```
